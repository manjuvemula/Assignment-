**Basic Python**
Task 1: Write a Python program to check if a number is even or odd.
The program should take an integer input from the user and print whether the number is even or odd.
Example: If the user inputs 4, the output should be "4 is even".

```python
number = int(input("Enter an integer: "))

if number % 2 == 0:
    print(f"{number} is even")
else:
    print(f"{number} is odd")
```

Task:2 Write a Python program to find the largest of three numbers. The program should take three integer inputs from the user and print the largest number. Example: If the user inputs 3, 7, and 5, the output should be "The largest number is 7"

```python
# Take three integer inputs from the user
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
num3 = int(input("Enter the third number: "))
largest = max(num1, num2, num3)
print(f"The largest number is {largest}")
```

**NumPy**
Task1: Create a NumPy array of shape (3, 3) filled with random numbers.
Then, compute the mean, median, and standard deviation of the array.

```python
import numpy as np
array = np.random.rand(3, 3)
mean = np.mean(array)
median = np.median(array)
std_dev = np.std(array)
print("Array:")
print(array)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```

Task 2: Perform matrix multiplication using NumPy.
Create two 2x2 matrices and multiply them. Print the result.


```python
import numpy as np
matrix1 = np.array([[1, 2],
                    [3, 4]])
print("Matrix 1:")
print(matrix1)

matrix2 = np.array([[5, 6],
                    [7, 8]])
result = np.dot(matrix1, matrix2)
print("\nMatrix 2:")
print(matrix2)
print("\nResult of matrix multiplication:")
print(result)
```

Task 2: Filter the DataFrame based on a condition.
For example, filter rows where a specific column value is greater than a threshold. Display the filtered
DataFrame.

```python
import pandas as pd

# Define threshold
threshold = 10

# Create DataFrame
df = pd.DataFrame({
    'city': ['A','B','C','D'],
    'one': [7,11,20,8],
    'two': [11,12,21,6],
    'three': [13,13,22,4],
    'four': [16,14,23,5],
})

# Filter DataFrame based on threshold
filtered_df = df[(df[['one', 'two', 'three']] > threshold).all(axis=1)]

print(filtered_df)

```

**Matplotlib**
Task 1: Create a line plot using Matplotlib.
Plot the function y = x^2 for x values ranging from -10 to 10. Label the axes and give the plot a title.


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = x**2
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^2')
plt.show()
```

**Seaborn**
Task 1: Create a histogram using Seaborn.
Use a sample dataset (e.g., tips dataset) to plot the distribution of a numerical column.

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.histplot(tips['total_bill'], kde=False, bins=10)
plt.xlabel('Total Bill Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Total Bill Amount')
plt.show()
```

Task 2: Create a heatmap using Seaborn.
Use a correlation matrix of a dataset to create a heatmap. Annotate the heatmap for better readability.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'A': [11, 12, 13, 14, 15],
    'B': [15, 14, 13, 12, 11],
    'C': [11, 12, 13, 14, 15],
    'D': [15, 14, 13, 12, 11],
    'E': [11, 12, 13, 14, 15]
}
df = pd.DataFrame(data)
corr_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
